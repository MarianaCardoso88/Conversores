```python
import os
import pandas as pd

# Diretório onde estão os arquivos txt e onde os arquivos xlsx serão criados
diretorio_txt = 'txt'

# Listar os arquivos txt na pasta
arquivos_txt = [arquivo for arquivo in os.listdir(diretorio_txt) if arquivo.endswith('.txt')]

# Para cada arquivo txt, crie um arquivo xlsx vazio com o mesmo nome
for arquivo_txt in arquivos_txt:
    nome_arquivo_xlsx = os.path.splitext(arquivo_txt)[0] + '.xlsx'  # Nome do arquivo xlsx correspondente
    caminho_arquivo_xlsx = os.path.join(diretorio_txt, nome_arquivo_xlsx)  # Caminho completo para o novo arquivo xlsx

    # Criar um DataFrame vazio e salvar como um arquivo xlsx
    df = pd.DataFrame()
    df.to_excel(caminho_arquivo_xlsx, index=False)

    print(f'Arquivo {nome_arquivo_xlsx} criado com sucesso.')

print('Todos os arquivos foram processados com sucesso.')
```
